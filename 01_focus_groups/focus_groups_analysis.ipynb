## <b><font color='white'> Analysis Main Facts

```python
sum_row_index = 18
```

```python
list_of_dataframes = [(df_tools_used, '1.1 Tools used'), 
                      (df_tools_most_used, '1.2 Most used'), 
                      (df_tasks, '1.3 Tasks'), 
                      (df_shortcuts, '1.4 Shortcuts used'), 
                      (df_add_shortcuts, '1.5 Shortcuts wished for'),
                      (df_work_pos, '1.6 Work Position'),
                      (df_seat_pos, '1.7 Seated Postion')
                      ]  


for subdf, df_name in list_of_dataframes:
    
    sum_row = subdf.iloc[sum_row_index]

    sum_row_sum = sum_row.sum()
    
    sum_count = pd.concat([sum_row, pd.Series({'Sum': sum_row_sum})])
    num_diff_results = len(sum_count)-1

    # Print the information
    print(df_name)
    print(sum_count)
    print(f'Count of different values: {num_diff_results}')
    print('\n')
```

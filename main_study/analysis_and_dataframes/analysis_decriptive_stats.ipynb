Extracting info from filenames 
- timestamp when the software was started

```python
def extract_infos(log_file):
    parts_filename = log_file.split('_')
    id = parts_filename[0]
    start_date = parts_filename[1]
    start_time = parts_filename[2]
    return id, start_date, start_time
```

Timestamp formatting

```python
def format_timestamp(date, time):
    dt_str = f"{date} {time}"
    dt_obj = datetime.strptime(dt_str, "%Y%m%d %H%M%S")
    return dt_obj
```

Metadata

```python
def metadata_dataframe(path):
    metadata = []
    for log_file in os.listdir(path):
        if log_file.endswith('.csv'):
            id, start_date, start_time = extract_infos(log_file)
            start = format_timestamp(start_date, start_time)
            
            metadata.append([
                log_file,
                id, 
                start_date,
                start,
            ])

    return pd.DataFrame(metadata, columns=['filename', 'ID', 'start_date', 'start_time'])
```
